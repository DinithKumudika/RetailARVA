```python
MODELS = [
    'bge-large',
    'mxbai-embed-large',
    'bge-m3',
    'nomic-embed-text',
    'snowflake-arctic-embed'
]
DATASET_SPLIT = 'test'
BATCH_SIZE = 32
```

```python
import ollama
from datasets import load_dataset
from scipy.stats import pearsonr
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import time
from tqdm import tqdm

# Load and prepare dataset
print("Loading STS dataset...")
dataset = load_dataset('sentence-transformers/stsb', split=DATASET_SPLIT)
```

```python
sentences = set()
pairs = []
scores = []
```

```python
for example in dataset:
    sentences.add(example['sentence1'])
    sentences.add(example['sentence2'])
    pairs.append((example['sentence1'], example['sentence2']))
    scores.append(example['score'])
```

```python
unique_sentences = list(sentences)
print(f"Loaded {len(unique_sentences)} unique sentences from {len(pairs)} pairs")
```

```python
# Benchmarking function
def benchmark_model(model_name):

    print(f"\nBenchmarking {model_name}...")

    embeddings = {}
    start_time = time.time()

    # Process each sentence individually
    for text in tqdm(unique_sentences, desc="Generating embeddings"):
        try:
            response = ollama.embeddings(
                model=model_name,
                prompt=text  # Single string instead of list
            )
            embeddings[text] = np.array(response['embedding'])
        except Exception as e:
            print(f"Error generating embeddings for {model_name}: {e}")
            return None

    embedding_time = time.time() - start_time
    embeddings_per_sec = len(unique_sentences) / embedding_time

    print(f"\nBenchmarking {model_name}...")

    # Similarity calculation
    predicted = []
    for s1, s2 in tqdm(pairs, desc="Calculating similarities"):
        emb1 = embeddings[s1]
        emb2 = embeddings[s2]
        sim = cosine_similarity([emb1], [emb2])[0][0]
        predicted.append(sim)

    # Calculate Pearson correlation
    correlation, _ = pearsonr(predicted, scores)

    return {
        'model': model_name,
        'correlation': correlation,
        'embeddings_per_sec': embeddings_per_sec,
        'embedding_time': embedding_time
    }
```

```python
# Run benchmarks
results = []
for model in MODELS:
    result = benchmark_model(model)
    if result:
        results.append(result)
```

```python
# Display results
print("\nBenchmark Results:")
print(f"{'Model':<25} | {'Correlation':<12} | {'Embed/s':<10} | {'Time (s)'}")
for res in results:
    print(f"{res['model']:<25} | "
          f"{res['correlation']:.4f}       | "
          f"{res['embeddings_per_sec']:,.2f}    | "
          f"{res['embedding_time']:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert results to DataFrame
df = pd.DataFrame(results)

# Set seaborn style
sns.set(style="whitegrid")

# 1. Correlation Comparison (Accuracy)
plt.figure(figsize=(8, 5))
sns.barplot(
    x='correlation',
    y='model',
    data=df.sort_values('correlation', ascending=False),
    palette='viridis'
)
plt.title('Model Accuracy Comparison (Pearson Correlation)', fontsize=14)
plt.xlabel('Correlation Score', fontsize=12)
plt.ylabel('Model', fontsize=12)
plt.tight_layout()
plt.show()

# 2. Speed Comparison (Embeddings per Second)
plt.figure(figsize=(8, 5))
sns.barplot(
    x='embeddings_per_sec',
    y='model',
    data=df.sort_values('embeddings_per_sec', ascending=True),
    palette='magma'
)
plt.title('Embedding Speed Comparison', fontsize=14)
plt.xlabel('Embeddings per Second', fontsize=12)
plt.ylabel('Model', fontsize=12)
plt.tight_layout()
plt.show()

# 3. Accuracy vs Speed Trade-off (Scatter Plot)
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='correlation',
    y='embeddings_per_sec',
    hue='model',
    data=df,
    s=100,
    palette='tab10',
    legend='full'
)
plt.title('Accuracy vs Speed Trade-off', fontsize=14)
plt.xlabel('Correlation Score', fontsize=12)
plt.ylabel('Embeddings per Second', fontsize=12)

# Add model names as annotations
for i, row in df.iterrows():
    plt.text(row['correlation'],
             row['embeddings_per_sec'],
             row['model'],
             fontsize=9,
             ha='center')

plt.tight_layout()
plt.legend(title='Models', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
